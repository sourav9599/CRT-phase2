```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>

struct Node{
	int Data;
	struct Node *Next,*Prev;
}*S,*L,*P,*C;


void insertNode();
void deleteNode();
void display();

int main(){
	int x;
	S = L = P = C = NULL;
	while(1){
		system("cls");
		printf("\n ------------------- ");
		printf("\n Doubly Linked List  ");
		printf("\n 1. Insert ");
		printf("\n 2. Delete ");
		printf("\n 3. Display ");
		printf("\n 4. Exit ");
		printf("\n Enter the Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: insertNode();
			break;
			case 2: deleteNode();
			break;
			case 3: display();
			break;
			case 4: printf("\n Program is Terminating");
			return 0;
			break;
		}
	}
	return 0;
}
void insertNode(){
	struct Node *nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter the value ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	nn->Prev = NULL;
	if(S==NULL){
		S = L = nn;
	}
	else{
		if(nn->Data < S->Data){
			nn->Next = S;
			S->Prev = nn;
			S = nn;
		}
		else if(nn->Data > L->Data){
			L->Next = nn;
			nn->Prev = L;
			L = nn;
		}
		else{
			for(C=S;C->Data < nn->Data;P=C,C=C->Next);
			nn->Next = C;
			nn->Prev = P;
			P->Next = nn;
			C->Prev = nn;
		}
	}
	printf("\n Node is Inserted...");
	getch();
}
void deleteNode(){
	if(S==NULL){
		printf("\n Doubly Linked List Empty");
	}
	else if(S->Next==NULL){
		S = L = NULL;
	}
	else{
		P = L->Prev;
		C = L;
		L = P;
		P->Next = NULL;
		C->Prev = NULL;
		C = NULL;
		P= NULL;
	}
	printf("Node is deleted...");
	getch();
}
void display(){
	if(S==NULL){
		printf("\n Doubly Linked List Empty");
	}
	else{
		int o;
		printf("\n Select the Order ");
		printf("\n 1. Fwd Order ");
		printf("\n 2. Rev order ");
		printf("\n Select the Order ");
		scanf("%d",&o);
		switch(o){
			case 1: printf("\n Doubly Linked List ");
					printf("\n ------------------ \n");
					for(C=S;C!=NULL;C=C->Next){
						printf("%d ",C->Data);
					}
			break;
			case 2:
					printf("\n Doubly Linked List ");
					printf("\n ------------------ \n");
					for(C=L;C!=NULL;C=C->Prev){
						printf("%d ",C->Data);
					}
			break;
		}
	}
	getch();
}
```
