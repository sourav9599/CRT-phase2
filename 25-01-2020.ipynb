```python
QUEUE (ARRAY):-
================

** One Direction for enter an item -- REAR
** Another Direction to remove an item -- FRONT
** Item will be removed always from FRONT and
items will enters into queue fro mthe REAR

** FIFO -- FIRST IN FIRST OUT
** ENQUEUE -- INSERT ITEM TO THE QUEUE
** DEQUEUE -- REMOVE ITEM FROM THE QUEUE
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
int main()
{
	int a[5],F=-1,R=-1,x,i;
	while(1){
		system("cls");
		printf("\n -------------- ");
		printf("\n ----1. INSERT  - ");
		printf("\n ----2. REMOVE  - ");
		printf("\n ----3. DISPLAY - ");
		printf("\n ----4. EXIT    - ");
		printf("\n Enter your Choice ");
		scanf("%d",&x);
		switch(x){
			case 1: if(R<4)
					{
						R++;
						if(R==0){
							F = 0;
						}
						printf("\n Enter a value ");
						scanf("%d",&a[R]);
						printf("\n Value is Inserted");
						getch();
					}
					else{
						printf("\n Queue is FULL");
						getch();
					}
			break;
			case 2: if(F != -1 && R != -1)
					{
						if(F == R)//F ==0 && R == 0
						{
							a[F] = 0;
							F--;
							R--;
							printf("\n Value is removed");
							getch();
						}
						else
						{
							for(i=F+1;i<=R;i++)
							{
								if(i==R)//Last time
								{
									a[i-1] = a[i];
									a[i] = 0;//a[R] = 0
									R--;//Decrease Index of R
								}
								else{
									a[i-1] = a[i];
								}
							}
							printf("\n Value is removed");
							getch();
						}
					}
					else
					{
						printf("\n Queue is Empty");
						getch();
					}
			break;
			case 3: if(F!=-1 && R != -1)
					{
						printf("\n ----------------- ");
						printf("\n --- Queue Values- ");
						for(i=0;i<=R;i++){
							printf("%d ",a[i]);
						}
						getch();
					}
					else{
						printf("\n Queue is Empty");
						getch();
					}
			break;
			case 4: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
```
