```python
Structures :-
-------------
1. Why Strctures?	

struct is keyword used to define the structure
Inside of the structure it can store different
datatype. It means you can store int,char,float
double and long

struct <structName>{
	Datatype MemberStruct;
	......................
};

** Structure defination does not have any memory
** This required the memory when the object or
variable of structure is created.

** Object Of Structure:-
1. Along with struct defination
	struct Demo{
		int Id;
	}obj1;
2. By using struct tag(inside main function)
	struct Demo obj;
```

```python
#include<stdio.h>
//Define a struct
struct Demo{
	int Id;
};
int main()
{
	struct Demo obj;//Creating the obj of struct
	obj.Id = 100;//. -- Data Binding Operator
	//It combine  the Data of two entities
	printf("%d",obj.Id);
	return 0;
}
```

```python
Bubble Sort:-
-------------
** It compares the adj items if the first item is greater than 
second item then swap the both the items

** Complex -- O(n^2)

-- Finding the first large or second large items of the list
then we can use Bubble Sort
```

```python
#include<stdio.h>
void bubbleSort(int a[],int n)
{
	int i,j,temp;
	for(i=0;i<2;i++){
		for(j=0;j<n-1-i;j++){
			if(a[j]>a[j+1]){
				temp = a[j];
				a[j] = a[j+1];
				a[j+1] = temp;
			}
		}
	}
	//printf("%d ",a[n-1]);
//	printf("%d ",a[n-2]);
	
}
int main()
{
	int x[6] = {19,2,25,18,12,5},i,n=6;
	bubbleSort(x,6);
	
	for(i=0;i<n;i++){
		printf("%d ",x[i]);
	}
	
	return 0;
}
```

```python
Binary Search:-
---------------

** This algorithm works only on Unique List
** This algorithm works only on Sorted List
** Complex - log(n)
** Best Case - O(1)
** Worst Case - log(n)
```

```python
#include<stdio.h>
int binarySearch(int a[],int leftIndex,int rightIndex,int tar)
{
	int mid;
	while(leftIndex<=rightIndex){
		mid = leftIndex + (rightIndex-leftIndex)/2;
		if(a[mid] == tar){
			return mid;
		}
		if(a[mid] < tar){
			leftIndex = mid + 1;
		}
		else{
			rightIndex =  mid - 1;
		}
	}
	return -1;
}
int main()
{
	int resIndex;
	int x[9] = {1,11,18,25,45,76,89,99,122};
	resIndex = binarySearch(x,0,8,121);
	if(resIndex>=0){
		printf("Item Found");
	}
	else{
		printf("Item is not Found");
	}
	return 0;
}
```

```python
#include<stdio.h>
int binarySearch(int a[],int n,int tar)
{
	int mid,leftIndex=0,rightIndex=n-1;
	while(leftIndex<=rightIndex){
		mid = leftIndex + (rightIndex-leftIndex)/2;
		if(a[mid] == tar){
			return mid;
		}
		if(a[mid] < tar){
			leftIndex = mid + 1;
		}
		else{
			rightIndex =  mid - 1;
		}
	}
	return -1;
}
int main()
{
	int resIndex;
	int x[9] = {1,11,18,25,45,76,89,99,122};
	resIndex = binarySearch(x,9,122);
	if(resIndex>=0){
		printf("Item Found");
	}
	else{
		printf("Item is not Found");
	}
	return 0;
}
```
