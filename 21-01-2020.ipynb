```python
#include<stdio.h>
int main()
{
	int a[10][10],r,c,i,j;
	//Reading the multi dimensional array
	scanf("%d%d",&r,&c);
	for(i=0;i<r;i++){
		for(j=0;j<c;j++){
			scanf("%d",&a[i][j]);
		}
	}
	//Printing of Multi Dimensional array
	for(i=0;i<r;i++){
		for(j=0;j<c;j++){
			printf("%d ",a[i][j]);
		}
		printf("\n");
	}
	return 0;
}
```
