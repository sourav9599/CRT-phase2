```python
Implementing STACK with LINKED LIST
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>

struct Node{
	int Data;
	struct Node *Next;
}*S,*P,*C,*Top;

void push();
void pop();
void display();

int main()
{
	int x;
	while(1){
		system("cls");
		printf("\n --------------- ");
		printf("\n ---- STACK ---- ");
		printf("\n ---- 1. PUSH -- ");
		printf("\n ---- 2. POP  -- ");
		printf("\n ---- 3. Display ");
		printf("\n ---- 4. Exit -- ");
		printf("\n Enter Your Choice ");
		scanf("%d",&x);
		switch(x){
			case 1: push();
			break;
			case 2: pop();
			break;
			case 3: display();
			break;
			case 4: printf("Program is Terminating");
			return 0;
			break;
		}
	}
	return 0;
}
void push(){
	struct Node *nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter a value ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(S == NULL){
		S = nn;
	}
	else{
		nn->Next = S;
		S = nn;
	}
	printf("\n Value is Pushed");
	getch();
}
void pop(){
	if(S==NULL){
		printf("\n Stack is Empty ");
		printf("\n Stack is Underflow ");
	}
	else{
		C = S;
		S = S->Next;
		C->Next = NULL;
		C = NULL;
	}
	printf("\n Node is deleted ...");
	getch();
}
void display(){
	if(S==NULL){
		printf("\n Stack is Empty ");
		printf("\n Stack is Underflow");
	}
	else{
		printf("\n Stack Values are : ");
		printf("\n ------------------- \n");
		for(C=S;C!=NULL;C=C->Next){
			printf("%d ",C->Data);
		}
	}
	getch();
}
```

```python
Implementing QUEUE with LINKED LIST
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>

struct Node{
	int Data;
	struct Node *Next;
}*F,*R,*C;

void insert();
void removeItem();
void display();

int main(){
	int x;
	while(1){
		system("cls");
		printf("\n ------------------- ");
		printf("\n ----- QUEUE ------- ");
		printf("\n ----- 1. Insert --- ");
		printf("\n ----- 2. Remove --- ");
		printf("\n ----- 3. Display -- ");
		printf("\n ----- 4. Exit  ---- ");
		printf("\n Enter the Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: insert();
			break;
			case 2: removeItem();
			break;
			case 3: display();
			break;
			case 4: printf("\n Program is Terminatin");
			return 0;
			break;
		}
	}
	return 0;	
}
void insert(){
	struct Node *nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter the Value : ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(F==NULL){
		F = R = nn;
	}
	else{
		R->Next = nn;
		R = nn;
	}
	printf("\n Value is Inserted...");
	getch();
}
void removeItem(){
	if(F==NULL){
		printf("\n QUEUE is Empty");
	}
	else if(F->Next == NULL)
	{
		F = R = NULL;
		printf("\n Value is Removed...");
	}
	else{
		C = F;
		F = F->Next;
		C->Next = NULL;
		C = NULL;
		printf("\n Value is Removed ...");
	}
	getch();
}
void display(){
	if(F==NULL){
		printf("\n QUEUE is Empty");
	}
	else{
		printf("\n QUEUE values are : ");
		printf("\n ----------------   \n");
		for(C=F;C!=NULL;C=C->Next){
			printf("%d ",C->Data);
		}
	}
	getch();
}
```
