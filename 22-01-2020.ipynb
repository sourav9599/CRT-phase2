```python
#include<stdio.h>
int main()
{
	int uCnt=0,lCnt=0,len = 0,i;
	char x[100];
	scanf("%[^\n]s",x);
	for(i=0;x[i]!='\0';i++){
		len++;
	}
	for(i=0;i<len;i++){
		if(x[i] >= 65 && x[i] <= 90){
			uCnt++;
		}
		else if(x[i] >= 97 && x[i] <= 122){
			lCnt++;
		}
	}
	printf("U : %d L : %d",uCnt,lCnt);
	return 0;
}
```

```python
#include<stdio.h>
#include<string.h>
int main()
{
	int k,s=0,len,i;
	char x[100];
	scanf("%[^\n]s",x);
	len = strlen(x);
	for(i=0;i<len;i++){
		if(x[i]>=48 && x[i]<=57){
			k = x[i] - 48;//Ascii converts to decimal
			if(k%2==0){
				s += k;
			}
		}
	}
	printf("%d",s);
	return 0;
}
```

```python
#include<stdio.h>
#include<string.h>
int main()
{
	int len,i,currentCnt=0,prevCnt=0,startIndex,endIndex;
	char x[100];
	scanf("%[^\n]s",x);
	len = strlen(x);
	//first or middel
	for(i=0;i<len;i++){
		if(x[i]!=32){
			currentCnt++;
		}
		else if(x[i]==32){
			if(prevCnt<currentCnt){
				prevCnt = currentCnt;
				startIndex = i - prevCnt;
				endIndex = i;
			}
			currentCnt = 0;
		}
	}
	//Last word verification
	if(prevCnt<currentCnt){
		prevCnt = currentCnt;
		startIndex = i - prevCnt;
		endIndex = i;
	}
	for(i=startIndex;i<=endIndex;i++){
		printf("%c",x[i]);
	}
	return 0;
}
```

```python
#include<stdio.h>
#include<string.h>
int main()
{
	int len,i,j,k;
	char x[100],ch='!';
	scanf("%[^\n]",x);
	len = strlen(x);
	for(i=0;i<len;i++){
		if(x[i]!=32){
			k = x[i]-64;//1 to 26
			for(j=0;j<k;j++){
				printf("%c",ch);
			}
			if(ch=='!'){
				ch = '@';
			}
			else{
				ch = '!';
			}
		}
		else{
			printf("\n");
			ch = '!';
		}
	}
	return 0;
}
```

```python
#include<stdio.h>
#include<string.h>
int main()
{
	int len,i,j,k;
	char x[100],ch='!';
	scanf("%[^\n]",x);
	len = strlen(x);
	for(i=0;i<len;i++){
		if(x[i]!=32){
			k = x[i]-64;//1 to 26
			for(j=0;j<k;j++){
				printf("%c",ch);
			}
			if(ch=='!'){
				ch = '@';
			}
			else if(ch=='@')
			{
				ch = '$';
			}
			else if(ch=='$'){
				ch = '!';
			}
		}
		else{
			printf("\n");
			ch = '!';
		}
	}
	return 0;
}
```
